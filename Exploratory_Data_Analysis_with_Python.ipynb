* ### Handling String '..' charaters
Most of the columns have '..' strings in them. Therefore needs to be process.

     - First replace with NAs
     - Convert to numeric data types for filling NUll values
     - Fill Null values with mean


```python
# List of columns to process
columns_to_process = ['GII VALUE', 'GII RANK','Maternal_mortality', 'Seats_parliamentt(% held by women)', 'F_secondary_educ', 'M_secondary_educ', 'F_Labour_force', 'M_Labour_force']

for column in columns_to_process:
    df[column] = df[column].replace('..', np.nan)

```

```python
# Removing commas from Maternal_mortality column
df['Maternal_mortality'] = df['Maternal_mortality'].str.replace(',', '')
```

```python
df
```

Now all '..' replaced with NaN

```python
# List of columns to convert to float
columns_to_convert = [
    'GII VALUE',
    'GII RANK',
    'Maternal_mortality',
    'Adolescent_birth_rate',
    'Seats_parliamentt(% held by women)',
    'F_secondary_educ',
    'M_secondary_educ',
    'F_Labour_force',
    'M_Labour_force'
]

# Replace non-numeric entries like 'NaN', '..', etc., with np.nan
df[columns_to_convert] = df[columns_to_convert].replace(['NaN', '..'], np.nan)

# Convert the selected columns to float
df[columns_to_convert] = df[columns_to_convert].astype(float)
```

```python
df
```

```python
# Count total NaN values in the entire DataFrame
total_nan_count = df.isna().sum().sum()
total_nan_count
```

That is a lot of NANs given the small amount of dataset.

So, I will fill NAN with mean or preceeding numbers

```python
# Fill NaN values with the mean and round to 2 decimal places
df = df.fillna(df.mean(numeric_only=True)).round(2)
df
```
